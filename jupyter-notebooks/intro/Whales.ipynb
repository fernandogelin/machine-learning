```python
data = gl.SFrame.read_csv("../whale_data/train.csv")
```

```python
path = '/Users/fernandogelin/Downloads/imgs/'
```

```python
def get_image(row):
    return gl.Image(path + row['image_name'])
```

```python
data['image'] = data.apply(get_image)
```

```python
train_data, test_data = data.random_split(.8, seed=0)
```

```python
train_data['image'].head().show()
```

```python
raw_pixel_model = gl.logistic_classifier.create(train_data,
                                                target='whale_id',
                                                features=['image'],
                                                validation_set = None)
```
